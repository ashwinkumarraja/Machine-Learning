```python
def readDataset(file):
    dataset = pd.read_csv(file)

    X = dataset.sample(frac = 1)

    X = np.array(X)

    X = np.append(np.ones((np.shape(X)[0],1)),X,1)
    label = dataset.iloc[:,-1:]

    X_train = np.array(X[:math.floor(len(X)*0.7)])
    X_test = np.array(X[math.floor(len(X)*0.7)+1:])

    print(X_train)
    print(X_test)
    
   
    
    
    return X_train,X_test,X
```

```python
def generateParams():
    #Seed
    w0 = np.random.uniform(-1,1,(np.shape(X)[1]-1,1))
    print("Seed \n",w0)
    numOfIterations  = 1000000
    print("Number of Iterations - ",numOfIterations)
    learning_rate = 1
#     print("Shape of Training Set " np.shape(X_train))
    
    return w0,numOfIterations,learning_rate
```

```python
def generatePerceptron(w0,numOfIterations,X_train,learning_rate,X):
    for f in range (numOfIterations):
        terminate = True

        misclassified = 0
        count = 0
        for i in X_train :
            count+=1
            x = np.reshape(i[:-1],(1,np.shape(X)[1]-1))

            lab = i[-1]
            if lab == 0:
                lab = -1

            w1 = np.matmul(x,w0)

            if ((lab)*w1 < 0):
                w0 = w0 + learning_rate*(np.transpose(x))*(lab)
                terminate = False
                misclassified+=1
                break
    
        if terminate :
            break
    #         print(w0)

    print("Perceptron :\n" ,w0)

    return w0
        
```

```python
def getAccuracy(w0,X_test,X):
    incorrect = 0
    for i in X_test:
        x = np.reshape(i[:-1],(1,np.shape(X)[1]-1))

        lab = i[-1]
        if lab == 0:
            lab = -1

        w1 = np.matmul(x,w0)
    #         print(w1)
        if ((lab)*w1 < 0):
            incorrect+=1


    print("Misclassified - ",incorrect)
    numRows = np.shape(X_test)[0]
    incorrect/= numRows
    print("Accuracy - ",1-incorrect)
    return
```

```python
print("\n \n Dataset 2 \n")
X_train,X_test,X = readDataset('dataset_LP_2.csv')
w0,numberOfIterations,learning_rate = generateParams();
w0 = generatePerceptron(w0,numberOfIterations,X_train,learning_rate,X)
accuracy = getAccuracy(w0,X_test,X)

print("Dataset 1 \n")
X_train,X_test,X = readDataset('dataset_LP_1.txt')
w0,numberOfIterations,learning_rate = generateParams();
w0 = generatePerceptron(w0,numberOfIterations,X_train,learning_rate,X)
accuracy = getAccuracy(w0,X_test,X)

```
